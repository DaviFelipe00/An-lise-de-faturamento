# Faturamento total de cada loja

```python
import pandas as pd
df = pd.read_excel(r'vendas.xlsx')
display(df.head(2))
```

```python
df_billing = df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
df_billing = df_billing.sort_values(by='Valor Final', ascending=False)
display(df_billing)
```

# Quantidade Vendida

```python
df_quantity = df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
df_quantity = df_quantity.sort_values(by='Quantidade', ascending=False)
display(df_quantity)
```

# Ticket Médio

```python
df_avarage_ticket = (df_billing['Valor Final'] / df_quantity['Quantidade']).to_frame('Ticket Médio') #to_frame = Função para criar nova coluna.
df_avarage_ticket = df_avarage_ticket.sort_values(by='Ticket Médio', ascending = False)
display(df_avarage_ticket)
```
